```python
###import libraries for web scraping
import requests
import lxml.html as lh
import pandas as pd
import numpy as np 
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
[len(T) for T in tr_elements[:12]]
```

```python
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print (i,name)
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 3, the //tr data is not from our table 
    if len(T)!=3:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
        
[len(C) for (title,C) in col]
```

```python
#place in dataframe
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
df.head()
```

```python
#Clean the cells with borough that is "Not assigned"
df['Borough'] = df['Borough'].replace(['Not assigned'],np.nan)
df
```

```python
#delete rows with missing data
df = df.dropna()
df
```

```python
#read postal_codes long,lat
dfCoor = pd.read_csv('http://cocl.us/Geospatial_data')
dfCoor.head(100)
```

```python
#rename Post Code to Postalcode 
dfCoor.rename(columns={'Postal Code':'Postcode'}, inplace=True)
dfCoor.dtypes
```

```python
#merge the dataframes base on Postcode
dfComplete = pd.merge(df, dfCoor, how='inner', on = 'Postcode')
dfComplete
```
